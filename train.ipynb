```python
from ultralytics import YOLO
import torch
import os

# Fix lỗi OpenMP hay gặp trên Windows
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"

print("Đang kiểm tra thiết bị...")
if torch.cuda.is_available():
    print(f"✅ Đã tìm thấy GPU: {torch.cuda.get_device_name(0)}")
    device = '0'
else:
    print("⚠️ Không tìm thấy GPU. Đang dùng CPU (Sẽ chậm).")
    device = 'cpu'
```

```python
# Tải model (Nếu máy chưa có nó sẽ tự tải, nếu có rồi nó dùng luôn)
model = YOLO('yolo11n.pt')
print("Đã tải model thành công!")
```

```python
yaml_path = 'D:/ThiGiacMayTinh/DemNguoi2/dataset/data.yaml'

# Bắt đầu train
results = model.train(
    data=yaml_path,
    epochs=100,          # Số vòng học
    imgsz=640,          # Kích thước ảnh
    device=device,      # Chạy bằng GPU
    batch=8,            
    workers=0,          
    project='KetQuaTrain',
    name='lan_train_so_1',
    plots=True          # Vẽ biểu đồ
)
```
